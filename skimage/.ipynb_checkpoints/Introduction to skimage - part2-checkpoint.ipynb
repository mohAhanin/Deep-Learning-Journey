```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
image = plt.imread('image.jpg')
plt.imshow(image)
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

# Image restoration

```python
def get_mask(image):
    mask = np.zeros(image.shape[:-1])
    mask[101:106, 0:240] = 1
    mask[152:154, 0:60] = 1
    mask[153:155, 60:100] = 1
    mask[154:156, 100:120] = 1
    mask[155:156, 120:140] = 1
    mask[212:217, 0:150] = 1
    mask[217:222, 150:256] = 1
    return mask
```

```python
from skimage.restoration import inpaint

# obtain the mask
mask = get_mask(image)

# Apply inpainting to the damaged image using the mask
restored_image = inpaint.inpaint_biharmonic(image, mask, channel_axis=-1)

# Show the resulting image
show_image(restored_image)
```

```python
show_image(mask)
```

```python
# Show the defect and resulting images
show_image(image,'Image to restore')
show_image(restored_image,'Image restored')
```
