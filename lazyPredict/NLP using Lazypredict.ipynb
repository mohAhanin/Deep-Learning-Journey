```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
```

```python
data = fetch_20newsgroups(subset='train', categories=['sci.space', 'comp.graphics'])
X = data.data
y = data.target
```

```python
X
```

```python
y
```

```python
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)
```

```python
X_tfidf_dense = X_tfidf.toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_dense, y, test_size=0.2, random_state=0)
```

```python
# Initialize LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models
```

```python
for model_name, model in clf.models.items():
    print(f"Model: {model_name}")
    print(f"Parameters: {model.get_params()}")
    print("\n")
```
